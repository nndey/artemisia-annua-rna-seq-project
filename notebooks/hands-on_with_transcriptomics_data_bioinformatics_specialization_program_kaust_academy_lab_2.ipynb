## About Artemisinin

Artemisinin is a compound that was first identified and extracted from the plant, _Artemisia annua_ L.  (i.e., sweet wormwood). The compound is known for its anti-plasmodic properties, making it an excellent treatment for malaria. As of 2018, therapies oriented around artemisinin were the gold standard as far as malaria treatment is concerned (Dondorp et al. 2009). 

Also in 2018, there was growth in research demonstrating the compound having anti-tumor, anti-parasitic, anti-fibrosis, and anti-arrhythmic properties (Bin & Hong 2010). These properties have increased demand for the compound due to its potential for treating a wider variety of diseases and ailments (Peplow 2013). 

However, given that artemisinin can only be extracted directly from _Artemisia annua_ L, which already has low quantities of the compound in most varieties, there is a major bottleneck to scaling production of the compound. While there have been attempts to chemically synthesize, synthesize via cell culture, and heterologously synthesize via _Escherichia coli_ or yeast (Paddon et al. 2013) on artemisinin, they too pose barriers to being viable avenues for scaled production.

As such, the main focus has been to increase/improve artemisinin content in Artemisia annua. Research investigating the impact on abiotic factors, particularly light, has been promising with UV B and C radiation improving artemisinin content in A. annua despite these spectra harming overall growth and development (Rai et al. 2011). 

For that reason, the study from which this portfolio project originated from opted to explore other light spectra and found that red and blue light enhanced artemisinin content by upregulating expression of genes involved in artemisinin biosynthesis (e.g., ADS and CYP71AV1).

```bash
# Count total bases for blue light samples (SRR6808227, SRR6808228, SRR6808226) using awk only
BlueSamples=("SRR6808226" "SRR6808227" "SRR6808228")

function LoopTotalBases
{
m=${#BlueSamples[@]}
# comment out m if you decide to not use indexing method
total_bases=0

for (( i=0; i<m; i++ ));
# can also do for file in "${BlueSamples[@]}";
do
   # define paired-end reads
   R1="${BlueSamples[$i]}_1.fastq"
   R2="${BlueSamples[$i]}_2.fastq"
   
   # matching logic wherein bases are only added after checking the number of bases in read 1 matches that of read 2
   # count number of bases in read 1
   bases_r1=$(awk 'NR%4==2 {sum+=length($0)} END {print sum}' $R1)
   
   # count number of bases in read 2
   bases_r2=$(awk 'NR%4==2 {sum+=length($0)} END {print sum}' $R2)
   
   # compare number of bases of read 1 to read 2
   # additional logic for cases where the paired end read counts don't match
   # if the paired end read counts dont match (the difference is not zero), take the read count of the read with lower number of bases
   # better to take read with lower number of bases
   if [ "$bases_r1" -eq "$bases_r2" ]; then
     # add current number of bases to total
     # can also do total_bases=$((total_bases + bases)
     ((total_bases+=bases_r1))
   elif [ "$bases_r1" -gt "$bases_r2" ]; then
       ((total_bases+=bases_r2))
   elif [ "$bases_r1" -lt "$bases_r2" ]; then
       ((total_bases+=bases_r1))
   else
       :
   fi
done

echo "Total bases for blue light samples: $total_bases"
}

LoopTotalBases ${BlueSamples[@]}

```
