## Explore GTF format (gene annotation)

### View header comments and first few data lines

```bash
head -50 GCA_003112345.1_ASM311234v1_genomic.gtf
```

### Count different feature types

```bash
echo "Feature types in GTF:"
cut -f3 GCA_003112345.1_ASM311234v1_genomic.gtf | grep -v "^#" | sort | uniq -c | sort -rn
```

### Count genes within genome

```bash
echo "Number of genes:"
awk -F'\t' '$3 == "gene"' GCA_003112345.1_ASM311234v1_genomic.gtf | wc -l
```

### Find a specific gene

```bash
grep "gene_id \"CTI12\"" GCA_003112345.1_ASM311234v1_genomic.gtf | head -5 
```

```bash
awk '
$3=="gene"{
  len=$5-$4+1
  print $0 "\t" len
}' GCA_003112345.1_ASM311234v1_genomic.gtf | sort -k10,10nr | head -1
```
